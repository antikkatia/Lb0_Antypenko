- **Культурні пам'ятки**:

```python
from IPython.display import SVG, display
import math

# Розрахунок координат семикутної зірки
def calculate_star_points(cx, cy, outer_radius, inner_radius, num_points=7):
    points = []
    angle = math.pi / num_points  # Кут між вершинами
    for i in range(2 * num_points):
        r = outer_radius if i % 2 == 0 else inner_radius
        x = cx + r * math.cos(i * angle)
        y = cy - r * math.sin(i * angle)
        points.append(f"{x},{y}")
    return " ".join(points)

# Координати центру зірки, радіуси та кількість вершин
cx, cy = 30, 75  # Центр зірки (всередині трикутника)
outer_radius = 15
inner_radius = 7

# Генерація SVG
svg_code = f"""
<svg width="300" height="150">
  <!-- Чорна смуга -->
  <rect width="300" height="50" fill="black" />
  <!-- Біла смуга -->
  <rect y="50" width="300" height="50" fill="white" />
  <!-- Зелена смуга -->
  <rect y="100" width="300" height="50" fill="green" />
  <!-- Червоний трикутник -->
  <polygon points="0,0 100,75 0,150" fill="red" />
  <!-- Семикутна біла зірка -->
  <polygon points="{calculate_star_points(cx, cy, outer_radius, inner_radius)}" fill="white" />
</svg>
"""

# Відображення прапора
display(SVG(data=svg_code))
```
